```python
# Setup — ejecuta esta celda primero (Shift+Enter)

def chequear_quiz(respuestas, correctas_y_temas):
    """Valida el quiz completo y devuelve feedback indicando que tema repasar."""
    aciertos = 0
    total = len(correctas_y_temas)
    fallos = []
    sin_responder = []

    for clave in sorted(correctas_y_temas.keys()):
        esperada, tema = correctas_y_temas[clave]
        recibida = respuestas.get(clave, 0)

        if recibida == 0:
            sin_responder.append(clave)
        elif recibida == esperada:
            aciertos += 1
        else:
            fallos.append((clave, tema))

    print('=' * 65)
    porcentaje = (aciertos / total) * 100
    print(f'  RESULTADO: {aciertos}/{total} correctas ({porcentaje:.0f}%)')
    print('=' * 65)

    if sin_responder:
        print()
        print(f'AVISO: tienes preguntas sin responder: {sin_responder}')
        print('Reemplaza los ceros y vuelve a ejecutar la validacion.')

    if fallos:
        print()
        print('Preguntas falladas — temas a repasar:')
        for clave, tema in fallos:
            print(f'  - Pregunta {clave}: revisar {tema}')

    print()
    if porcentaje >= 70 and not sin_responder:
        print('APROBADO. Estas listo(a) para participar de Clase 1.')
    elif sin_responder:
        print('PENDIENTE. Termina de responder antes de evaluar tu resultado.')
    else:
        minimo = int(0.7 * total) if (0.7 * total).is_integer() else int(0.7 * total) + 1
        print(f'NO APROBADO. Necesitas minimo {minimo} respuestas correctas (70%).')
        print('Repasa los temas indicados en la guia y reintenta el quiz.')

print('Setup completo. Continua con las preguntas.')
```
